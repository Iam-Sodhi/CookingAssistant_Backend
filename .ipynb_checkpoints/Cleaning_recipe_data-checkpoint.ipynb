```python
import json

# Step 1: Load the nested recipe data
with open('recipe_data.json', 'r', encoding='utf-8') as file:
    recipe_data = json.load(file)

# Step 2: Define a function to clean the ingredient data
def clean_ingredient(ingredient):
    # Phrases that incorrectly appear as "Ingredient"
    invalid_ingredient_phrases = ["as per taste", "as required", "as per use", "a pinch","as per your taste", "as per required"]
    
    # If "Ingredient" is an invalid phrase, replace it with the "Unit"
    if ingredient["Ingredient"].lower() in invalid_ingredient_phrases:
        print(ingredient)
        ingredient["Ingredient"] = ingredient["Unit"]  # Move the correct ingredient name
        ingredient["Unit"] = ""  # Clear the unit since it's misplaced
        ingredient["Quantity"] = "Not specified"  # Set quantity as unknown
        print(ingredient)
    
    return ingredient

# Step 3: Clean each recipe's ingredients
for recipe, data in recipe_data.items():
    data["Ingredients"] = [clean_ingredient(ingredient) for ingredient in data["Ingredients"]]

# Step 4: Save the cleaned data back to a JSON file
with open('cleaned_recipe_data.json', 'w', encoding='utf-8') as outfile:
    json.dump(recipe_data, outfile, indent=4)

print("Data cleaned and saved to 'cleaned_recipe_data.json'.")
```
